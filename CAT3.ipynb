## What is World Happiness Report?
According to the world happiness index, the World Happiness Report is an indicator of global happiness. The rankings are based on reports of respondents' assessments of their own lives, and articles on national happiness are included. The report also compares with other elements that influence (quality of) life.


## ABOUT THE DATASET

The happiness scores and rankings use data from the Gallup World Poll. The scores are based on answers to the main life evaluation question asked in the poll. This question, known as the Cantril ladder, asks respondents to think of a ladder with the best possible life for them being a 10 and the worst possible life being a 0 and to rate their own current lives on that scale. The scores are from nationally representative samples for the years 2013-2016 and use the Gallup weights to make the estimates representative. 

The columns following the happiness score estimate the extent to which each of six factors – economic production in form of GDP per Capita, social support, life expectancy, freedom, absence of corruption, and generosity – contribute to making life evaluations higher in each country than they are in Dystopia, a hypothetical country that has values equal to the world’s lowest national averages for each of the six factors. They have no impact on the total score reported for each country, but they do explain why some countries rank higher than others.

## REGRESSION ANALYSIS

```python

import pandas as pd
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm

x=whi[['Explained_by:_GDP_per_capita','Explained_by:_Social_support','Explained_by:_Healthy_life_expectancy',
       'Explained_by:_Freedom_to_make_life_choices','Explained_by:_Generosity','Explained_by:_Perceptions_of_corruption']]
y=whi['Happiness_score']
regr = LinearRegression()
regr.fit(x, y)
```

```python
print('Intercept: \n',regr.intercept_)
print("----------------------------------------")
print('Coefficients: \n', regr.coef_)
```

## INTERPRETATION
The multiple linear regression model is-

HAPPINESS SCORE = 5.133 - 0.08277692  * GDP_per_capita   + 0.06987223 * Social_support  -0.01212464  * Healthy_life_expectancy - 0.00685409 * Freedom_to_make_life_choices   + 0.01265533 * Generosity - 0.01740167 * Perceptions_of_corruption 

### Analysis of the intercept
When every factors explaining the happiness score in the model is zero then the happiness score of a country will be 5.133

### Analysis of the coefficients

With 1 per capita increase in GDP the happiness score decreases by 0.08277692 <br>
With 1 unit increase in Social support the happiness score increase by 0.06987223 <br>
With 1 unit increase in Healthy life expectancy the happiness score decreases by 0.01212464 <br>
With 1 unit increase in Freedom to make life choices the happiness score decreases by 0.00685409 <br>
With 1 unit increase in Generosity the happiness score increases by 0.01265533 <br>
With 1 unit increase in Perceptions of corruption the happiness score decreases by 0.01740167 <br>
    

```python
# TO PREDICT THE HAPPINESS SCORE FOR THE THREE COUNTRIES
predictions = regr.predict(x)
print(predictions)
```
